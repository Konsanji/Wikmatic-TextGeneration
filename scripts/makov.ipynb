```python
with open("UN_news100.txt") as f:
    text = f.read()
```

```python
# créer modèles de chaines de markov
text_model = markovify.Text(text, state_size=2, well_formed=True)
```

```python
print(text_model.make_sentence())
```

```python
for i in range(200):
    text_model = markovify.Text(text, state_size=2, well_formed=True)
    print(text_model.make_sentence(tries=50, test_output=True))
```

```python
for i in range(200):
    text_model = markovify.Text(text, state_size=1, well_formed=False)
    print(text_model.make_short_sentence(max_chars=50, tries=100))
```

```python
text_model.make_sentence_with_start('Today', state_size=2)
```

```python
with open("UN2.txt") as f:
    text2 = f.read()
```

```python
for i in range(200):
    text_model = markovify.Text(text2, state_size=1, well_formed=True)
    print(text_model.make_sentence(tries=200, test_output=True))
```

```python
text1_model = markovify.Text(text, state_size=1, well_formed=True)
text2_model = markovify.Text(text2, state_size=1, well_formed=True)
```

```python
combined_model = markovify.combine([text1_model, text2_model], [1,1]) 
```

```python
print(combined_model.make_sentence())
```
